# Compare between-modality spcorrs to within-modality reliabilities

```python
files = glob.glob('spcorr/spcorrs_summary_*.csv')
for file in files:
    roi = '_'.join(file.split('/')[-1].split('.')[0].split('_')[2:])
    data = pd.read_csv(file)
    within_READ = data.SN_SN.values
    within_LIST = data.ID_ID.values
    between = data.ID_SN_EE.values

    geometric_mean = np.sqrt(np.mean(within_READ)*np.mean(within_LIST))
    print(f'\n{roi} within reading = {np.mean(within_READ)}')
    print(f'{roi} within listening = {np.mean(within_LIST)}')
    print(f'{roi} between reading and listening = {np.mean(between)}')
    print(f'{roi} geometric mean = {geometric_mean}')
    stats = scipy.stats.wilcoxon(between-geometric_mean)
    print(f'{roi} p = {stats.pvalue} (one sample Wilcoxon signed rank test)')
```
